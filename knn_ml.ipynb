```python
from scipy.spatial import distance
import pandas as pd
import numpy as np
```

```python
Dataset = pd.DataFrame({'Brightness':[40,50,60,10,70,60,25],
           'Saturation' : [20, 50, 90, 25, 70, 10, 80],
           'Class' : ['Red', 'Blue', 'Blue', 'Red', 'Blue', 'Red', 'Blue']})
final_data = Dataset
final_data
```

```python
Distance=[]
dataset = np.array(Dataset.iloc[:])
Test_data = [['Brightness', 'Saturation', 'Class'], [20, 35,'?']]
x2,y2 = Test_data[1][0], Test_data[1][1]
print(x2, y2)
```

```python
for i in range(7):
  x1 = dataset[i][0]
  y1 = dataset[i][1]
  x = [x1,x2]
  y = [y1,y2]
  dis = distance.euclidean(x,y)
  Distance.append(dis)
print(Distance)
```

```python
Dataset['Distance'] = Distance
Data = Dataset.sort_values(by = ['Distance'],ascending = True, inplace= False, ignore_index = True)
k = int(input("Enter k value to divide and finalise class  "))
Data
```

```python
red, blue = 0,0
for i in range(k):
  if Data['Class'][i] == 'Red':
    red+=1
  else:
    blue += 1
if red>blue:
  print("Red is the class of new dataset")
  Test_data[1][2] = 'Red'
  Distance.append(Test_data)
elif blue>red:
  Test_data[1][2] = 'Blue'
  Distance.append(Test_data)
  print("Blue is the class of new dataset")
else:
  print("Enter different k value since red and blue values are same")
print(Test_data[1])
final_data.loc[len(Data.index)] = Test_data[1]
final_data
```
